```python
import pandas as pd
```

```python
csv_file1 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Pickup Five Cities Datasets\pickup_hz.csv'
csv_file2 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Pickup Five Cities Datasets\pickup_yt.csv'
csv_file3 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Pickup Five Cities Datasets\pickup_sh.csv'
csv_file4 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Pickup Five Cities Datasets\pickup_cq.csv'
csv_file5 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Pickup Five Cities Datasets\pickup_jl.csv'

csv_files = [csv_file1,csv_file2,csv_file3,csv_file4,csv_file5]

# Load all CSV files into dataframes
dfs = [pd.read_csv(file) for file in csv_files]

# Merge all dataframes into one
df_combined_pickup = pd.concat(dfs, ignore_index=True)
```

```python
csv_file6 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Delivery Five Cities Datasets\delivery_hz.csv'
csv_file7 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Delivery Five Cities Datasets\delivery_yt.csv'
csv_file8 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Delivery Five Cities Datasets\delivery_sh.csv'
csv_file9 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Delivery Five Cities Datasets\delivery_cq.csv'
csv_file10 = r'C:\Users\Dusty\Downloads\Internship\Last-Mile-Delivery-Delays-and-Route-Optimization\data\Delivery Five Cities Datasets\delivery_jl.csv'

csv_files_delivery = [csv_file6,csv_file7,csv_file8,csv_file9,csv_file10]

# Load all CSV files into dataframes
dfs_delivery = [pd.read_csv(file_delivery) for file_delivery in csv_files_delivery]

# Merge all dataframes into one
df_combined_delivery = pd.concat(dfs_delivery, ignore_index=True)
```
